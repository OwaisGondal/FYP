### Myshop

```python
url = "https://myshop.pk/mobiles-smartphones-tablets/smartphones?a_product_type=969&product_list_limit=64&product_list_order=name"

driver.get(url)
for i in range(1, 64):
    try:
        Name = driver.find_element_by_xpath(
            "/html/body/div[1]/main/div[3]/div[1]/div[3]/ol/li["
            + str(i)
            + "]/div/div[2]/strong/a"
        )
        Price = driver.find_element_by_xpath(
            "/html/body/div[1]/main/div[3]/div[1]/div[3]/ol/li["
            + str(i)
            + "]/div/div[2]/div[4]/span/span/span[2]/span"
        )
        Names.append(Name.text)
        Prices.append(Price.text)
        Links.append(Name.get_attribute("href"))
        Vendors.append(V[3])
    except:
        1 == 1
```

### Zaibis

```python
url = "https://www.zaibis.com/product-category/new-mobile-phones/"

driver.get(url)
for j in range(1, 12):
    driver.get(
        "https://www.zaibis.com/product-category/new-mobile-phones/page/" + str(j) + "/"
    )
    for i in range(1, 32):
        #         Name = driver.find_element_by_xpath(
        #             "/html/body/div[2]/div/div[3]/div/div/div[1]/main/ul/li["
        #             + str(i)
        #             + "]/div/div/div[1]/a/h2"

        #         )
        try:
            Name = driver.find_element_by_xpath(
                '//*[@id="main"]/ul/li[' + str(i) + "]/div/div/div[1]/a"
            )
            Price = driver.find_element_by_xpath(
                '//*[@id="main"]/ul/li[' + str(i) + "]/div/div/div[3]/div[1]/span"
            )
            Names.append(Name.text)
            Prices.append(Price.text)
            Vendors.append(V[2])
            if Name.get_attribute("href") != "None":
                Links.append(Name.get_attribute("href"))
        except:
            1 == 1
```

### HomeShopping

```python
url = "https://homeshopping.pk/categories/Mobile-Phones-Price-Pakistan"
# driver = webdriver.Chrome('C:/Users/Usman/Downloads/chromedriver.exe')
driver.get(url)
for j in range(1, 21):
    driver.execute_script("window.scrollTo(0,document.body.scrollHeight-1200)")
    time.sleep(3)
    for i in range(1, 25):
        try:
            Name = driver.find_element_by_xpath(
                "/html/body/div[1]/div[9]/div[2]/div[2]/div[5]/div/div["
                + str(i * j)
                + "]/div/h5/a"
            )
            Price = driver.find_element_by_xpath(
                "/html/body/div[1]/div[9]/div[2]/div[2]/div[5]/div/div["
                + str(i * j)
                + "]/div/div[2]/a/div"
            )
            Names.append(Name.text)
            Prices.append(Price.text)
            Vendors.append(V[5])
            Links.append(Name.get_attribute("href"))
        except:
            1 == 1
```
