```python
import bs4 as bs
import re
import pandas as pd
from selenium import webdriver
name=[]
price=[]
url = 'https://homeshopping.pk/categories/Mobile-Phones-Price-Pakistan'

driver = webdriver.Chrome('C:/Users/Usman/Downloads/chromedriver.exe')
driver.get(url)
for j in range(1,21):
    driver.execute_script("window.scrollTo(0, 300)")
    ++j
    for i in range(1,25):
        target_name=driver.find_element_by_xpath('/html/body/div[1]/div[9]/div[2]/div[2]/div[5]/div/div['+str(i)+']/div/h5/a')
        target_prize=driver.find_element_by_xpath('/html/body/div[1]/div[9]/div[2]/div[2]/div[5]/div/div['+str(i)+']/div/div[2]/a/div')
        name.append(target_name.text)
        price.append(target_prize.text)

driver.close()
```

```python
df=pd.DataFrame(list(zip(name,price)),columns=["Name","Price"])
df.sort_values(["Name","Price"],inplace=True)
df.reset_index(0,inplace=True)
df
```

```python
import bs4 as bs
import re
import pandas as pd
from selenium import webdriver
name=[]
price=[]
url = 'https://myshop.pk/mobiles-smartphones-tablets/smartphones?a_product_type=969&product_list_limit=64&product_list_order=name'

driver = webdriver.Chrome('C:/Users/Usman/Downloads/chromedriver.exe')
driver.get(url)
for i in range(1,64):
    if i==8 or i==10 or i==11 or i==14 or i==17 or i==20 or i==58:
        ++i
    elif i>60:
        target_name=driver.find_element_by_xpath('/html/body/div[1]/main/div[3]/div[1]/div[3]/ol/li['+str(i)+']/div/div[2]/strong/a')
        target_prize=driver.find_element_by_xpath('/html/body/div[1]/main/div[3]/div[1]/div[3]/ol/li['+str(i)+']/div/div[2]/div[4]/span/span/span[2]/span')
        name.append(target_name.text)
        price.append(target_prize.text)
    else:
        target_name=driver.find_element_by_xpath('/html/body/div[1]/main/div[3]/div[1]/div[3]/ol/li['+str(i)+']/div/div[2]/strong/a')
        target_prize=driver.find_element_by_xpath('/html/body/div[1]/main/div[3]/div[1]/div[3]/ol/li['+str(i)+']/div/div[2]/div[4]/span/span/span')
        name.append(target_name.text)
        price.append(target_prize.text)

driver.close()
df=pd.DataFrame(list(zip(name,price)),columns=["Name","Price"])
df.sort_values(["Name","Price"],inplace=True)
df.reset_index(0,inplace=True)
df
```
