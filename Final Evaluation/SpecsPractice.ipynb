#### Scraping 

```python
BrandKeys = []
BrandValues = []
for i in range(1, 35):
    BrandKeys.append(
        driver.find_element_by_xpath(
            '//*[@id="container"]/div[3]/div/div[2]/section[2]/ul/li[' + str(i) + "]/a"
        ).text
    )
    BrandValues.append(
        driver.find_element_by_xpath(
            '//*[@id="container"]/div[3]/div/div[2]/section[2]/ul/li[' + str(i) + "]/a"
        ).get_attribute("href")
    )
```

```python
Brands = dict(zip(BrandKeys, BrandValues))
Phones = dict()
Phone = None

for key in Brands:
    Phones[key.split(" ")[0]] = {}
# Phones
# Brands
```
